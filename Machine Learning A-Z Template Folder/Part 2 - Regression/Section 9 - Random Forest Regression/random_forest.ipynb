```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
```

```python
dataset=pd.read_csv("Position_Salaries.csv")
X=dataset.iloc[:,1:2].values
y=dataset.iloc[:,2].values

from sklearn.ensemble import RandomForestRegressor

regrassor=RandomForestRegressor(n_estimators=300,random_state=0)
regrassor.fit(X,y)

regrassor.predict(6.5)
```

```python
X_grid=np.arange(min(X),max(X),0.001)
X_grid.reshape(len(X_grid),1)
plt.figure(figsize=(50,10))
plt.scatter(X,y,color="red")
plt.plot(X,regrassor.predict(X),color="blue")
plt.title("Trurth or bluff decision tree regression")
plt.xlabel("No of years of experience")
plt.ylabel("Salary")
plt.show()

```
