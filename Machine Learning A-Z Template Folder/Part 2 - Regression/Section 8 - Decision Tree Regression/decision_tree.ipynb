```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv("Position_Salaries.csv")
X=dataset.iloc[:,1:2].values
y=dataset.iloc[:,2]
```

```python
from sklearn.tree import DecisionTreeRegressor

regraasor=DecisionTreeRegressor(random_state=0)
regraasor.fit(X,y)
y_predict=regraasor.predict(6.5)
y_predict
```

```python

X_grid=np.arange(min(X),max(X),1)
X_grid.reshape(len(X_grid),1)
plt.scatter(X,y,color="red")
plt.plot(X,regraasor.predict(X),color="blue")
plt.title("Trurth or bluff decision tree regression")
plt.xlabel("as")
plt.ylabel("xsd")

plt.show()

```
